##Stemming

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stemmer = PorterStemmer()
stemmSent = []
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words("english"))]
  stemmSent.append(' '.join(words))
```

```python
stemmSent
```

##Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmetizer = WordNetLemmatizer()
lemmeSent = []
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [lemmetizer.lemmatize(word) for word in words if word not in set(stopwords.words("english"))]
  lemmeSent.append(' '.join(words))
```

```python
lemmeSent
```

##Bag Of Words

```python
import re
```

```python
corpus = []
```

```python
for i in range(len(sentences)):
  review = re.sub('[^a-zA-Z]', ' ', sentences[i])
  review = review.lower()
  review = review.split()
  review = [lemmetizer.lemmatize(word) for word in review if word not in set(stopwords.words("english"))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

##Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
vector = cv.fit_transform(corpus).toarray()
```

```python
vector
```

```python
for i in range(len(vector)):
  print(vector[i,:])
```
